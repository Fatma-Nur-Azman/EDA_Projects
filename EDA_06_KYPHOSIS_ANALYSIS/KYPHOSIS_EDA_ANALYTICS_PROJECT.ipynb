<a id="project-description"></a>
<h2 align="left"><font color=#FF0000> Project Description:</font></h2>   
    
    
- In this project, we will perform basic Exploratory Data Analysis (EDA) on the Kyphosis disease Dataset. 
- Kyphosis is an abnormally excessive convex curvature of the spine,  and 
- the dataset provides information on children who have undergone corrective spinal surgery. 
 <a id="about-the-dataset"></a>   
<h2 align="left"><font color=#FF0000> About the Dataset </font></h2>
    
- Dataset contains 81 rows and 4 columns representing data on children who have had corrective spinal surgery.
<br>    
- INPUTS: 
    1. **Age**: The age of the patient in months., 
    2. **Number**: The number of vertebrae involved in the surgery, 
    3. **Start**: The number of the first (top-most) vertebra operated on.
    4. **Kyphosis**: Indicates whether kyphosis was present (yes/no) after the operation. 

    <br>      
- OUTPUTS: Represents a factor with levels "absent" or "present," indicating if **kyphosis** (a type of deformation) was present after the operation.
    

<a id="getting-to-know-about-anatomy-of-vertebrae"></a>
<h3 align="left"><font color=#FF0000>  Getting to Know About Anatomy of Vertebrea  </font></h3>

**Anatomic Vertebrea Groups**

1. **Cervical Spine (C1-C7)** (Boyun)
- Number of Vertebrea: 7

2. **Thoracic Spine (T1-T12)** (Sırt)
- Number of Vertebrea: 12

3. **Lumbar Spine (L1-L5)** (Bel)
- Number of Vertebrea: 5

4. **Sacrum (S1-S5, fused)** (Kuyruk)
- Number of Vertebrea: 5

5. **Coccyx (Co1-Co4, fused)** (Kuyruk)
- Number of Vertebrea: 4

These groups represent the different sections of the vertebral column, with each section containing a specific number of vertebrae.

<a id="classifying-the-start-column-to-anatomic-vertebra-groups"></a>
<h3 align="left"><font color=#FF0000> Classifying the 'Start' column to Anatomic Vertebrea Groups </font></h3>

In this Dataset; **Start** column is the **first** (top-most) vertebra number operated on and containes the numbers between (1-18).

- We categorized the vertebra numbers in the Start column into Anatomical groups, and
- new column 'VertebraGroup'  is created to classify the values in the Start column.

**Vertebrea Groups for the 'Start' column**
- 1-7: Cervical (C1-C7) (boyun)
- 8-19: Thoracic (T1-T12) (sirt)
- 20-24: Lumbar (L1-L5) (bel)
- 25-29: Sacrum (S1-S5) (kuyruk)


<br>

- We labeled each group of vertebreas with a number and created new column called 'VertebraGroupNumber'.

**Labels of the Vertebrea Groups**
- (1) Cervical (C1-C7)
- (2) Thoracic (T1-T12)
- (3) Lumbar (L1-L5)
- (4) Sacrum (S1-S5)
- (5) Unknown or other

```python
# Convert the age column datatype from int64 to float64

# Yaş sütununun veri tipini int64'ten float64'e dönüştürelim
df['Age'] = df['Age'].astype('float64')

# Dönüşümü kontrol etmek için yaş sütununun veri tipini ve ilk birkaç satırı görüntüleyelim
print(df.dtypes)
print(df.head())
```

```python
# Define a function that converts age from months to years
# Apply the function to the “Age” column and add the results into a new column entitled “Age in Years”

# Yaşı aylardan yıllara dönüştüren fonksiyonu tanımlayalım
def convert_age_months_to_years(age_in_months):
    """Convert age from months to years."""
    age_in_years = age_in_months / 12
    return age_in_years

# Age sütununu dönüştürelim
df['Age_in_years'] = df['Age'].apply(convert_age_months_to_years)

# İlk birkaç satırı ve veri tiplerini kontrol edelim
print(df[['Age', 'Age_in_years']].head())
print(df.dtypes)
```

# <font color= #FF0000> <b> What are the mean, minimum, and maximum values of the age distribution?</b><font color= #FF0000>

```python
# List the average, minimum and maximum age (in years) considered in this study using 2 methods

import pandas as pd

# Convert the Age column to years (1 year = 12 months)
df['Age_in_years'] = df['Age'] / 12

# Calculate the average, minimum, and maximum  for age
average_age = df['Age_in_years'].mean()
minimum_age = df['Age_in_years'].min()
maximum_age = df['Age_in_years'].max()

print(f"Average Age: {average_age:.2f} years")
print(f"Minimum Age: {minimum_age:.2f} years")
print(f"Maximum Age: {maximum_age:.2f} years")
```

<a id="distribution-of-vertebra-groups"></a>
# <font color= #FF0000> <b> What is the distribution of vertebra groups?</b><font color= #FF0000>

```python
# Function to classify vertebra into Anatomic groups

def classify_vertebra(start):
    if 1 <= start <= 7:
        return 'Cervical (Boyun)'
    elif 8 <= start <= 19:
        return 'Thoracic (Sırt)'
    elif 20 <= start <= 24:
        return 'Lumbar (Bel)'
    elif 25 <= start <= 29:
        return 'Sacral (Sakral)'
    else:
        return 'Unknown'

# Apply the function to create a new column
df['VertebraGroup'] = df['Start'].apply(classify_vertebra)

# Display the unique vertebra groups and their counts
vertebra_groups = df['VertebraGroup'].value_counts()
print(vertebra_groups)


plt.figure(figsize=(10, 6))
vertebra_groups.plot(kind='bar', color='skyblue')
plt.title('Vertebra Groups Distribution')
plt.xlabel('Vertebra Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```


<font color= #FF0000>  Output:.</font>  
**Data Distribution in the Dataset**:

- The graph shows that the number of individuals in the thoracic (back) region is much higher than those in the cervical (neck) region in the dataset. This may indicate that the study focuses more on the thoracic region or that more data has been collected from this region.

**Research and Analysis**:
- The higher number of individuals in the thoracic (back) region can provide more information about kyphosis and other vertebral issues in this region. On the other hand, the lower amount of data from the cervical (neck) region may limit the conclusions that can be drawn about this area.

**Clinical Studies**:
- This distribution is important for clinical studies and treatment planning. The greater number of individuals in the thoracic region suggests that treatments and interventions may need to be more common in this area.
    

<a id="classification-with-numerical-labels"></a>
# <font color= #FF0000> <b> How are vertebra groups classified with numerical labels and what are the results of this classification?</b><font color= #FF0000>

```python
# Create a Function to Label of the VertebreaGroups values

def classify_vertebra(start):
    if 1 <= start <= 7:
        return 1  # Cervical (Boyun)
    elif 8 <= start <= 19:
        return 2  # Thoracic (Sırt)
    elif 20 <= start <= 24:
        return 3  # Lumbar (Bel)
    elif 25 <= start <= 29:
        return 4  # Sacral (Sakral)
    else:
        return 5  # Unknown or other

# Create a new 'VertebreaGroupNumbers' column and apply the classification function
df['VertebraGroupNumbers'] = df['Start'].apply(classify_vertebra)

# Check the results
print(df[['Start', 'VertebraGroup', 'VertebraGroupNumbers']].head())
```

<a id="distribution-of-kyphosis-by-vertebra-groups"></a>
# <font color= #FF0000> <b> What is the distribution of kyphosis by vertebra groups?</b><font color= #FF0000>

```python
# Group and count the kyphosis conditions by vertebraGroup
kyphosis_by_vertebra = df.groupby('VertebraGroup')['Kyphosis'].value_counts(normalize=True).unstack().fillna(0)

# Reset index to convert the DataFrame for Plotly
kyphosis_by_vertebra = kyphosis_by_vertebra.reset_index()
kyphosis_by_vertebra.columns = ['VertebraGroup', 'Absent', 'Present']

# Convert to percentage
kyphosis_by_vertebra['Absent'] *= 100
kyphosis_by_vertebra['Present'] *= 100

# Plot using Plotly with specified colors
fig = px.bar(kyphosis_by_vertebra, 
             x='VertebraGroup', 
             y=['Absent', 'Present'], 
             title='Range of Kyphosis by the VertebraGroup', 
             labels={'value': 'Percentage', 'VertebraGroup': 'Vertebra Groups'},
             barmode='stack',
             color_discrete_map={'Absent': 'skyblue', 'Present': 'darkblue'})

fig.update_layout(legend_title_text='Kyphosis')

# Show the plot
fig.show()
```


<font color= #FF0000>  Output:  Group and count the kyphosis conditions by vertebraGroup.</font>  
    
**Cervical (Neck) Group**:
- Kyphosis Absent: In this group, the proportion of individuals without kyphosis is approximately 75%. This indicates that the likelihood of kyphosis occurring in the cervical vertebrae is low.
- Kyphosis Present: In this group, the proportion of individuals with kyphosis is approximately 25%. This also implies that the likelihood of kyphosis occurring in the cervical vertebrae is lower compared to those without kyphosis.

**Thoracic (Back) Group**:
- Kyphosis Absent: In this group, the proportion of individuals without kyphosis is approximately 65%. This indicates that the likelihood of kyphosis occurring in the thoracic vertebrae is low, but slightly higher compared to the cervical vertebrae.
- Kyphosis Present: In this group, the proportion of individuals with kyphosis is approximately 35%. This implies that the likelihood of kyphosis occurring in the thoracic vertebrae is higher compared to the cervical vertebrae.


<a id="recurrence-status-of-kyphosis"></a>
# <font color= #FF0000> <b>  What is the recurrence status of kyphosis by vertebra number?</b><font color= #FF0000>

```python
# Group and count the kyphosis conditions by vertebra number
kyphosis_by_vertebra = df.groupby('Start')['Kyphosis'].value_counts(normalize=True).unstack().fillna(0)

# Reset index to convert the DataFrame for Plotly
kyphosis_by_vertebra = kyphosis_by_vertebra.reset_index()

# Rename columns for better readability in the plot
kyphosis_by_vertebra.columns = ['Start', 'Absent', 'Present']

# Convert to percentage
kyphosis_by_vertebra['Absent'] *= 100
kyphosis_by_vertebra['Present'] *= 100

# Plot using Plotly with specified colors
fig = px.bar(kyphosis_by_vertebra, 
             x='Start', 
             y=['Absent', 'Present'], 
             title='Range of Kyphosis by the Vertebra Numbers', 
             labels={'value': 'Percentage', 'Start': 'Start (Vertebra Numbers)'},
             barmode='stack',
             color_discrete_map={'Absent': 'skyblue', 'Present': 'darkblue'})

# Update x-axis to ensure labels are shown correctly
fig.update_layout(
    xaxis=dict(
        tickmode='linear',  # ensures all ticks are shown
        tick0=1,            # start at the first tick
        dtick=1             # interval between ticks
    ),
    legend_title_text='Kyphosis'
)

# Show the plot
fig.show()
```


<font color= #FF0000>  Output:  Plot showing the frequency of the Start vertebra number:.</font>  

**Distribution of Kyphosis by Vertebra Numbers**:
- In most vertebra numbers, the "Absent" (not present) condition of kyphosis is predominant. This indicates that kyphosis is less frequently observed in most vertebra numbers.
- Particularly in vertebra numbers 3, 6, 12, and 14, the "Present" (present) condition of kyphosis is more noticeable. This suggests a higher likelihood of kyphosis being present in these vertebra numbers.

**Notably Significant Vertebrae**:
- Vertebra 3 and 6: In these vertebrae, the "Present" (present) condition of kyphosis constitutes a significant percentage. This indicates that kyphosis may be more common in these vertebra numbers.
- Vertebra 12 and 14: Similarly, in these vertebrae, the "Present" (present) condition of kyphosis is also significant. This further indicates a higher likelihood of kyphosis occurring in these vertebra numbers.

**General Occurrence of Kyphosis**:
- In most vertebra numbers, the "Absent" (not present) condition of kyphosis is predominant. This suggests that kyphosis is generally rare across vertebra numbers.
- Since kyphosis is more frequently present in certain vertebrae (3, 6, 12, 14), patients with issues in these areas may need closer examination.

# <font color= #FF0000> <b> CORRELATION ANALYSIS </b><font color= #FF0000>

```python
# Sadece sayısal verileri içeren bir alt veri çerçevesi oluştur
numerical_df = df.select_dtypes(include=['number'])

# HeatMap: Korelasyon matrisi
correlation_matrix = numerical_df.corr()

# Korelasyon matrisi oluştur
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of All Numerical Variables')
plt.show()
```

<font color= #FF0000>  Output:  CORRELATION ANALYSIS :.</font>     
    
In the graph, we see the correlation coefficients between the columns “Age”, “Flumber” and “Start”.
Correlation between Age and Number (number of vertebrae):
    The correlation coefficient is -0.02. This means that there is almost no correlation between age and the number of vertebrae operated on.
    
**Correlation between Age and Start:**
    
The correlation coefficient is 0.06. This indicates that there is a very weak positive correlation between age and the number of vertebrae where the surgery started.
    
**Correlation between Number (number of vertebrae) and Start (starting vertebra):**
    
The correlation coefficient is -0.41. This indicates that there is a moderate negative correlation between the number of vertebrae operated on and the number of vertebrae where the surgery started. That is, as the number of vertebrae operated on increases, the number of vertebrae where the surgery started tends to decrease.


    
    
   **Conclusion**
    
    
**If we look at the general results of the correlation matrix**:
The variable “Age” has weak correlations with the variables “Number” and “Start”. This indicates that age has no significant effect on the number of vertebrae operated on or the number of vertebrae on which the surgery started.
There is a moderate negative correlation between “Number” and “Start”. This indicates that as the number of vertebrae operated on increases, the surgery starts at lower levels.
This correlation analysis helps us to understand the basic relationships between variables and provides a basis for further analysis. For example, we can look deeper into how these variables interact to predict the presence of kyphosis after surgery.

## <font color= #FF0000> <b> What is the relationship between age, number, start data, and kyphosis? </b><font color= #FF0000>


```python
# İlgili sütunları seçelim ve 'Kyphosis' sütununu binary değere dönüştürelim
df['Kyphosis'] = df['Kyphosis'].apply(lambda x: 1 if x == 'present' else 0)

# İlişkili değişkenleri seçelim
variables = ['Age', 'Number', 'Start']

# Grafik düzenini belirleyelim (2x2 düzeninde)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Scatterplot - Age vs Number
sns.scatterplot(x='Age', y='Number', hue='Kyphosis', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Age vs Number')

# Scatterplot - Age vs Start
sns.scatterplot(x='Age', y='Start', hue='Kyphosis', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Age vs Start')

# Scatterplot - Number vs Start
sns.scatterplot(x='Number', y='Start', hue='Kyphosis', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Number vs Start')

# Pairplot diyagonal histogramlar için - Kyphosis'e göre Age histogramı
sns.histplot(df, x='Age', hue='Kyphosis', element='step', ax=axes[1, 1], kde=True)
axes[1, 1].set_title('Kyphosis vs Age Histogram')

# Genel başlık
fig.suptitle('Kyphosis vs Age, Number, Start', y=1.02)
plt.tight_layout()
plt.show()
```
